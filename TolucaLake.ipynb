```python
import pandas as pd
```

```python
df = pd.read_csv('https://data.lacity.org/api/views/pvft-t768/rows.csv?accessType=DOWNLOAD')
df['total'] = 1
```

```python
toluca_df = df[df.NCName == 'GREATER TOLUCA LAKE NC']
```

```python
toluca_df.head(2)
```

```python
mobile_open_pending = toluca_df[(toluca_df.RequestSource == 'Mobile App') & (toluca_df.Status == 'Open' ) | (toluca_df.Status == 'Pending' )  ]
```

```python
import folium

colors = {
    'Pending': 'blue',
    'Open': 'red'
}

m = folium.Map(location=[34.1541456,-118.3621504], zoom_start=15, zoomControl=False)


for idx, row  in mobile_open_pending.iterrows():
    tool_tip = '{}: {}'.format(row.Status, row.RequestType)
    icon = folium.Icon(color=colors[row.Status], icon='info-sign')
    info_tip = row.UpdatedDate
    folium.Marker([row['Latitude'], row['Longitude']], popup=info_tip, tooltip=tool_tip, icon=icon).add_to(m)
    
```
